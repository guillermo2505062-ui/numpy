```python
import random
```

```python
def generar_matriz(N):
    matriz= []
    for i in range (N):
        fila = []
        for j in range(N):
            fila.append(random.randint(1, 9))
        matriz.append(fila)
    return matriz

N = 5

matriz_aleatoria = generar_matriz(N)
for fila in matriz_aleatoria:
    print(fila)

def imprimir_matriz(matriz):
    for fila in matriz:
        fila_formateada = ' '.join(f'{elemento:2}' for elemento in fila)
        print(fila_formateada)
```

```python
print("Matriz de 5x5")
m_5 = generar_matriz(5)
imprimir_matriz(m_5)
print("============")
print("Matriz de 10x10")
m_10 = generar_matriz(10)
imprimir_matriz(m_10)
print("============")
print("Matriz de 20x20")
m_20 = generar_matriz(20)
imprimir_matriz(m_20)
```
